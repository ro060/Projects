## Task 2:- Explore football skills and cluster football players based on their attributes.


```python
dat=data[['short_name','age','height_cm','weight_kg','overall','potential','value_eur','wage_eur','weak_foot','skill_moves']]
```

```python
#Here, we are taking few players on ovr becuase if we take whole data cluster will be clumsy
dat = dat[dat.overall > 86]
dat
```

```python
names = dat.short_name.tolist() # saving names for later
dat = dat.drop(['short_name'], axis = 1) # drop the short_name column 
```

```python
from sklearn import preprocessing

x = dat.values # numpy array
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

## Use PCA to reduce 10 columns into 2

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) # 2D PCA for the plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

## Perform K-Means clustering


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)# specify the number of clusters
kmeans = kmeans.fit(reduced)# fit the input data
labels = kmeans.predict(reduced)# get the cluster labels
centroid = kmeans.cluster_centers_# centroid values
clusters = kmeans.labels_.tolist()# cluster values
```

## Make a new data frame by adding players' names and their cluster


```python
reduced['cluster'] = clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name']
reduced.head()
```

```python
sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='cluster', data = reduced, legend=False,
                   fit_reg=False, size = 15, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(reduced.x, reduced.y, reduced.name):
    texts.append(plt.text(x, y, s))

ax.set(ylim=(-1, 1))
plt.tick_params(labelsize=15)
plt.xlabel("PC 1", fontsize = 15)
plt.ylabel("PC 2", fontsize = 15)

plt.show()
```
